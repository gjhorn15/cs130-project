```python
import pandas as pd

url = "https://corgis-edu.github.io/corgis/datasets/csv/video_games/video_games.csv"

# Load dataset
df = pd.read_csv(url)

# Show first few rows to understand structure
print(df.head())

# Basic shape of the dataset
print("Rows, Columns:", df.shape)

# Summary statistics for numerical fields
print(df.describe())

# Count of unique consoles
print("Unique Consoles:", df["Release.Console"].nunique())

# Number of games per year
games_per_year = df["Release.Year"].value_counts().sort_index()
print("Games by Year:\n", games_per_year)
```

```python
# Notebook 2: Sales vs Review Score Analysis

import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("https://corgis-edu.github.io/corgis/datasets/csv/video_games/video_games.csv")

# Scatter plot: Review Score vs Sales
plt.figure(figsize=(7,5))
plt.scatter(df["Metrics.Review Score"], df["Metrics.Sales"], alpha=0.6)
plt.title("Video Game Reviews vs Sales")
plt.xlabel("Review Score")
plt.ylabel("Sales (millions)")
plt.grid(True)
plt.show()

# Correlation between review score and sales
correlation = df["Metrics.Review Score"].corr(df["Metrics.Sales"])
print("Correlation (Review Score vs Sales):", correlation)
```

```python
# Notebook 3: Exploring Popular Genres

import pandas as pd

df = pd.read_csv("https://corgis-edu.github.io/corgis/datasets/csv/video_games/video_games.csv")

# Many games have multiple genres, split them
df["Genres_List"] = df["Metadata.Genres"].str.split(",")

# Explode genres so we count each genre
exploded = df.explode("Genres_List")
exploded["Genres_List"] = exploded["Genres_List"].str.strip()

# Top 10 genres by count
genre_counts = exploded["Genres_List"].value_counts().head(10)
print("Top Genres by Game Count:\n", genre_counts)

# Average review score by genre
avg_review_by_genre = exploded.groupby("Genres_List")["Metrics.Review Score"].mean().sort_values(ascending=False)
print("Avg Review Score by Genre:\n", avg_review_by_genre.head(10))
```

```python
# Notebook 4: Playtime (Length) Distribution

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("https://corgis-edu.github.io/corgis/datasets/csv/video_games/video_games.csv")

# Histogram of average play length
plt.figure(figsize=(8,5))
sns.histplot(df["Length.All PlayStyles.Average"].dropna(), bins=30, kde=True)
plt.title("Distribution of Average Playtime (All Playstyles)")
plt.xlabel("Hours")
plt.show()

# Compare main story vs completionist
plt.figure(figsize=(8,5))
sns.kdeplot(df["Length.Main Story.Average"].dropna(), label="Main Story")
sns.kdeplot(df["Length.Completionists.Average"].dropna(), label="Completionists")
plt.title("Main Story vs Completionist Play Times")
plt.xlabel("Hours")
plt.legend()
plt.show()

# Print mean values
print("Mean All Playstyles:", df["Length.All PlayStyles.Average"].mean())
print("Mean Main Story:", df["Length.Main Story.Average"].mean())
print("Mean Completionist:", df["Length.Completionists.Average"].mean())
```
